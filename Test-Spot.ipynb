```python
import shutil, pathlib

# Source: where KaggleHub stored your files
src = pathlib.Path("/home/datahub/.cache/kagglehub/datasets/rohanrao/formula-1-world-championship-1950-2020/versions/24")

# Destination: inside your project folder
dst = pathlib.Path("/home/datahub/files/Final-Project-Formula1/data/raw")
dst.mkdir(parents=True, exist_ok=True)

# Copy all CSVs
for file in src.glob("*.csv"):
    shutil.copy(file, dst / file.name)

print("✅ Files copied to:", dst)
print("Files:")
for f in dst.glob("*.csv"):
    print("-", f.name)
```
